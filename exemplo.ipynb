```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dt = pd.read_csv("/content/sample_data/PokemonData.csv")

dt.rename(columns={
    "Name": "Nome_Pokemon",
    "Type1": "Tipo_1",
    "Type2": "Tipo_2",
    "HP": "Vida",
    "Attack": "Ataque",
    "Defense": "Defesa",
    "SpAtk": "Ataque_Especial",
    "SpDef": "Defesa_Contra_Especial",
    "Speed": "Velocidade",
    "Generation": "Geracao",
    "Legendary": "Lendario"
    }, inplace=True)

pokemon_1_tipo = dt[dt['Tipo_2'].isna()]

pokemon_2_tipos = dt[dt['Tipo_2'].notna()]

contagem_tipos = {
    "1 Tipo": pokemon_1_tipo.shape[0],
    "2 Tipos": pokemon_2_tipos.shape[0]
}

plt.figure(figsize=(8, 6))
sns.barplot(x=list(contagem_tipos.keys()), y=list(contagem_tipos.values()), palette='Blues_d', hue=list(contagem_tipos.keys()), legend=False)

plt.title("Comparação entre Pokémon com 1 Tipo e 2 Tipos", fontsize=16)
plt.xlabel("Número de Tipos", fontsize=14)
plt.ylabel("Quantidade de Pokémon", fontsize=14)
plt.show()
```

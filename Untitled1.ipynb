```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/sample_data/netflix_titles.csv')
df.head()
```

```python
df.shape
df['type'].value_counts()
```

```python
df['country'].value_counts().head(10)
```

```python
df['release_year'].value_counts().head(10)
plt.figure(figsize=(10,5))
df['release_year'].value_counts().sort_index().plot()
plt.title("Conteúdos lançados por ano")
plt.xlabel("Ano")
plt.ylabel("Quantidade")
plt.show()
```

```python
df['rating'].value_counts()
```

```python
genres = df['listed_in'].str.split(', ', expand=True).stack()
genres.value_counts().head(10)
```

```python
movies = df[df['type'] == 'Movie'].copy()

movies['duration'] = movies['duration'].str.replace(' min','').astype(float)

movies['duration'].mean()

plt.figure(figsize=(8,5))
sns.histplot(movies['duration'], bins=30)
plt.title("Distribuição da duração dos filmes na Netflix")
plt.xlabel("Duração (minutos)")
plt.ylabel("Quantidade")
plt.show()
```

```python
movies['duration'].describe()
```

```python
content_year = df.groupby(['release_year','type']).size().unstack()

content_year.plot(figsize=(10,5))
plt.title("Filmes vs Séries ao longo dos anos")
plt.show()
```
